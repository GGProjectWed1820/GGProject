```python
import networkx as nx
import dataclasses
from src.vertex_params import VertexParams, VertexType
from src.visualization import draw
```

```python
from src.productions.production1 import Production1
from src.productions.production2 import Production2
from src.productions.production9 import Production9
from src.productions.production10 import Production10
from src.productions.production12 import Production12
```

```python
graph = nx.Graph()
graph.add_nodes_from(
    [(
        0,
        dataclasses.asdict(
            VertexParams(vertex_type=VertexType.START, position=(0.5, 0.5), level=0)
        ),
    )]
)
```

```python
draw(graph)
```

```python
prods = [Production1(), Production9(), Production9(), Production12(), Production2(), Production9()]#, Production10()]
levels = [1, 2, 2, 2, 3, 3]#, 3]
```

```python
for level, prod in zip(levels, prods):
    subgraph = prod.find_isomorphic_to_left_side(graph=graph)
    graph = prod.apply(graph=graph, subgraph=subgraph)
    draw(graph)
    draw(graph, level=level)
    print(graph)
```
